```python
from sklearn import datasets
import numpy as np
import pandas as pd
```

```python
X, y= datasets.make_blobs(n_samples=800, centers=6,n_features=2,random_state=1)
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(X[:, 0], X[:, 1])
```

```python
from sklearn.cluster import KMeans

estimator = KMeans(n_clusters=6, init='k-means++')
estimator.fit(X)
y_pred = estimator.predict(X)

plt.scatter(X[:, 0], X[:, 1], c=y_pred)
plt.show()
```

```python
from sklearn.metrics import silhouette_score 
score = silhouette_score(X, y_pred)
print("模型轮廓系数为(1 最好, -1 最差):", score)
```
