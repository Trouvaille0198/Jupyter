```python
import seaborn as sns
sns.set()
```

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.relplot(x='sepal_length',y='sepal_width',data=iris,hue='species')
```

```python
sns.relplot(kind='line',x='sepal_length',y='sepal_width',data=iris,hue='species')
```

```python
sns.catplot(x="sepal_length", y="species", data=iris)
```

```python
sns.catplot(x="sepal_length", y="species", kind="swarm", data=iris)
```

```python
sns.catplot(x="sepal_length", y="species", kind="box", data=iris)
```

```python
sns.catplot(x="sepal_length", y="species", kind="boxen", data=iris)
```

```python
sns.catplot(x="sepal_length", y="species", kind="point", data=iris)
```

```python
sns.catplot(x="sepal_length", y="species", kind="bar", data=iris)
```

```python
sns.catplot(x="species", kind="count", data=iris)
```

```python
sns.displot(x=iris["sepal_length"],kind='kde',data=iris)
```

```python
sns.displot(x=iris["sepal_length"],kind='hist',data=iris,hue='species')
```

```python
sns.displot(x=iris["sepal_length"],kde=True,data=iris)
```

```python
sns.set()
sns.displot(x=iris["sepal_length"],y=iris['sepal_width'],kind='kde',hue='species',data=iris,multiple='fill')
```
