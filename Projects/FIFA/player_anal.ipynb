```python
def position_classify(row):
    if(row['Position']=='ST'):
        return '中锋'
    elif(row['Position']=='CAM'):
        return '前腰'
    elif(row['Position']=='LW' or row['Position']=='RW' or row['Position']=='CF' or row['Position']=='LM' or row['Position']=='RM'):
        return '边锋'
    elif(row['Position']=='CM'):
        return '中场'
    elif(row['Position']=='CDM'):
        return '后腰'
    elif(row['Position']=='CB'):
        return '中后卫'
    elif(row['Position']=='LB' or row['Position']=='RB' or row['Position']=='LWB' or row['Position']=='RWB'):
        return '边后卫'
    else:
        return '?'
```

```python
s=data.apply(lambda row:position_classify(row),axis=1)
```

```python
data_simple=data[['Position','PACE','SHOOTING','PASSING','DRIBBLING','DEFENCE','PHYSICAL']]
```

```python
data['位置']=s
data_simple['位置']=s
```

```python
data_simple.to_csv('../Data/players_info/players_info_classified.csv',index=False,encoding='utf-8-sig')
```

```python
data_compare=data_simple.query("位置=='中锋' or 位置=='边锋' or 位置=='中场' or 位置=='中后卫' or 位置=='边后卫'")
```
